```python
n = int(input())
count = 0
check = True

for _ in range(n):
    word = input()    # abbca
    pre = ''
    s = set()

    for i in word:
        if i != pre:
            if i in s:
                check = False
                break
            s.add(i)
        pre = i

    if check:
        count += 1

print(count)
```
