```python
# bool의 쇼타임 시작됩니다.
# 주의할점 True x true , False x false

# 부등호
3 > 1 # 3이 1보다 크다 True
3 == 1 # 3은 1과 같다 False
20 != 5 # 20은 5와 같지 않다 True
20 != 20 # 20은 20과 같지 않다 False
```

```python
# 문자열 비교
"python" == "Python" # False -> 대소문자 구분됨.
"Python" == "Python" # True
"Python" != "python" # True
```

```python
# 중요
# ==, is

3 is 3
3 == 3

3.0 is 3
3.0 == 3

print(id(3.0))
print(id(3))
```

```python
# 본격적인 논리연산자
# True : 1 / False : 0

True and True # True // 1*1=1
True and False # False // 1*0=0
False and False # False // 0*0=0

True or True # True // 1+1=1
True or False # True // 1+0=1
False or False # False // 0+0=0

not True # False
not False # True
```

```python
# 점점 복잡해지는 조건들!!

10 == 23 or 6 != 2 # True
2 > 5 and 7 == 7 # False
not 14 <= 15 # False
not 7 is 7.0 # True
```
