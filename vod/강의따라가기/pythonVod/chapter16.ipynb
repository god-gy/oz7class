### while문 목차
1. while문 만들어보기
2. while문 코드 흐름 보기
3. while문 조건식 증감
4. while문을 조심스럽게 사용해야하는 이유
5. while문이 for문 보다 강점인 부분
6. continue, break (흐름 제어 키워드)
7. while문에도 else가 있다.
8. 붕어빵 인터네셔널
9. 예제로 while문 익혀보기

```python
# 1. while문 만들어보기

i = 0
while i < 10:
    print('점점 성장하는 여러부~~ㄴ', i)
    i += 1
```

```python
# 2. while문 코드 흐름 보기

i = 0                    # 초기식
while i < 10:            # while 조건식
    print('동물')         # 반복할 코드
    print('보러갈래요?')    # 반복할 코드
    i += 1               # 변환식
```

```python
# 3. while문 조건식 증감

# 1) 기본적인 증가 유형

i = 1
while i < 10:
    print('이정도로 성장할 줄이야',i,'증가')
    i+=1

# 2) 기본적인 감소
i = 10
while i > 0:
    print('이정도로 성장할 줄이야',i,'감소')
    i-=1
```

```python
# 4. while문을 조심스럽게 사용해야하는 이유
# 조건이 잘못되면 무한 루프에 빠져 과도한 리소스 소모
# 하지만, 잘 사용하면 for문보다 효율적임.

# 1 무한 출력
while True:
    print(1)
```

```python
# 5. while문이 for문 보다 강점인 부분
# 무수히 많이 실행해야할때

import random

# random.randint(1,6)

i = 0
while i !=4:
    i = random.randint(1,6)
    print(i)
```

```python
# 7. while문에도 else가 있다.

양파 = 10
while 양파 > 10:
    print("양파", 양파, "개 남았습니다.")
    if 양파 ==0:
        break
    양파 -=1
else:
    print("양파 다 팔렸습니다.")
```

```python
# 6. break (흐름 제어 키워드)

i = 0
while True:
    print(i)
    i+=1
    if i ==10:
        break

for i in range(100):
    print(i,"for")
    if i ==10:
        break
```

```python
# 6. continue

i = 0
while i < 10:
    i +=1
    if i % 2 == 0:
        continue
    print(i)

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# 8. 붕어빵 인터네셔널
# for문을 이용해 리팩토링 해주세요. input()
# 잡담 게시판에 코드 자랑 필수

# 초기 코드
fish__p , fish__s , fish__j = "붕어빵(팥)", "붕어빵(슈크림)", "붕어빵(잡채)"
fish__p__sell , fish__s__sell , fish__j_sell = 2000, 2500, 3000
order_s, order_j = 2, 3
order_s__pay, order_j__pay = fish__s__sell * 2, fish__j_sell *3
total = order_s + order_j
total_pay = order_s__pay + order_j__pay

print(f'주문이 완료되었습니다. {fish__s} {order_s}개, {fish__j}개 {order_j} 총 {total}개 결제 금액은 {total_pay}원 입니다.')
```

```python
# 9. 예제로 while문 익혀보기

# 콜라츠라는 사람은 주어진 수가 1이 될때까지 다음 작업을 반복하면, 모든 수를 1로 만들 수 있다고 추측했다.
# 작업
# 1-1. 입력된 수가 짝수라면 2로 나눈다.
# 1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더한다.
# 2. 위 결과로 나온 수에 같은 작업을 1이 될때까지 반복한다.
# 예를 들어 입력된 수가 6이라면 6->3->10->5->16->8->4->2->1

# num 변수에 있는 수를 1로 만드는데 몇번이나 작업을 반복해야하는지 출력한다.
num = int(input())

# 코드를 작성하세요.
count = 0                   # 반복 횟수 변수 설정
while num != 1:             # num 변수가 1이 아닌 것이 True라면 실행해라.
    if num % 2 == 0:        # num 변수를 나머지 없이 2로 나눈 것이 0과 같다면 (짝수라면)
        num = num // 2      # num 변수에 2로 나눈 값을 넣는다.
    else:                   # num 변수가 나머지 없이 2로 나눈 것이 0과 같지 않다면 (홀수라면)
        num = num * 3 + 1   # num 변수에 3을 곱하고 1을 더한 값을 넣는다.
    count += 1              # 반복시마다 count 변수에 1을 더한다.
print("최종",count)
```
