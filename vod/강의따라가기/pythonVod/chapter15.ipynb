```python
# 7. 예제로 for문 익혀보기
# 문제 : 절대값 정수가 담긴 정수와 그 정수의 부호를 차례로 담은 부호 변수가 있다
# 두 변수를 이용해 실제 정수의 합을 구하세요 (True:양수, False:음수)

정수 = [10,5,8]
부호 = [True, False, True]

합 = 0

for k, v in zip(부호, 정수):        # 부호와 정수를 짝지어서 각 변수 k와 v에 담고 반복한다.
    if k:                         # 만약 k가 양수라면
        합 += v                   # 합 변수에 v를 더한다.
    else:                         # 그게 아니라면 (k가 음수라면)
        합 -= v                   # 합 변수에 v를 뺀다 (음수에 빼기를 하면 더하기가 됨.)
print(f'총 합은 {합}입니다.')         # for문이 종료된 후 합 변수를 출력한다.
```
