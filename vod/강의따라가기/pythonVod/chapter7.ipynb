```python
생년월일 = "2000.12.25"

# 2000 12 25

print(생년월일[0:4]+" "+생년월일[5:7]+" "+생년월일[8:10])

# 파머 답안

년 = 생년월일[0:4]
월 = 생년월일[5:7]
일 = 생년월일[8:10]

print(f'{년} {월} {일}')
```

```python
숫자 = "123412413523142341251"
# 전체출력
print(숫자[::])

# 전체출력 (역순)
print(숫자[::-1])

# 첫번째 인덱스 기준으로 2칸씩 띄어쓰기 출력하기
print(숫자[::2])
```
