### list
* list란? 원소들이 연속적으로 저장되는 형태의 자료형을 말함.
* 시퀀스 자료형!!
    * str(정복), list(정복 예정), tuple, dict, set
* 만드는 방법은 쉬움. 대괄호로 묶어주고, 값들은 콤마로 구분함.

```python
# 첫 리스트 두근두근

리스트 = []
print(리스트)

리스트 = [23, "이게 리스트다", 3.14, True, False]
print(리스트)
```

```python
# 리스트 다시 한번 선언해 보자

리스트 = ["이", "렇", "게", "만", "들", "어", "요"]
리스트
리스트2 = list() # 형변환 int(), float() 같은 너낌
리스트2

# range -> 연속된 숫자를 생성하는 기능.

range(10)

# list(range(시작, 끝))
리스트_매직_들어간다 = list(range(0,20))
리스트_매직_들어간다
```

```python
# list(range(시작, 끝, 증가))
리스트_매직2 = list(range(0, 20, 2))
print(리스트_매직2)

리스트_매직3 = list(range(20,0,-1))
print(리스트_매직3)

리스트_매직4 = list(range(20,0))
print(리스트_매직4)
```

```python
# 슬라이싱 퀴즈

로또 = [3, 5, 15, 33, 41, 44]

# 로또 변수에 있는 리스트의 인덱스 1번부터 2번까지의 값을 출력.
print(로또[1:3])

# 로또 변수의 마지막 위치 출력.
print(로또[-1])

# 반대로 출력.
print(로또[::-1])

# 로또 리스트 안에 33이 있는지 확인.
print(33 in 로또)

# 로또 리스트를 로또2 리스트와 합치기.
로또2 = [2, 12, 15, 24, 33, 39]
print(로또 + 로또2, "로또 리스트를 로또2 리스트와 합치기.")

# range를 이용해서 1부터 10 사이에 짝수만 들어있는 짝수(변수명) 리스트를 만들기.
십까지 = list(range(1,11))
짝수1 = 십까지[1:11:2]
print(짝수1, "range를 이용해서 1부터 10 사이에 짝수만 들어있는 짝수(변수명) 리스트를 만들기.")

짝수 = list(range(2,12,2))
print(짝수, "짝수2")

# 짝수 리스트에 들어있는 값을 2배 늘리고 다시 짝수 변수에 담기.
짝수 = 짝수 * 2
print(짝수, "짝수 리스트에 들어있는 값을 2배 늘리고 다시 짝수 변수에 담기.")

# 짝수 리스트에 들어있는 요소의 개수를 구하라.
print(len(짝수), "짝수 리스트에 들어있는 요소의 개수를 구하라.")

# 짝수 리스트 3번째 인덱스를 출력.
print(짝수[3], "짝수 리스트 3번째 인덱스를 출력.")

# len() 함수를 이용하여 인덱스의 마지막 값을 출력.
print(len(짝수), "인덱스의 개수")
print(len(짝수)-1, "마지막 인덱스에 부여된 번호")
print(짝수[len(짝수)-1], "짝수 리스트의 마지막 인덱스 값")
```

```python
# list 값 넣고, 바꾸고, 지우고

mzfood = ["숙주", "분모자", "마라", "소세지", "소고기", "옥수수면"]
print(mzfood)

# 값을 넣는 방법 2가지 : append()->마지막 요소에 값을 넣음, insert()->원하는 곳에 값을 넣음.
mzfood.append("고수")
print(mzfood, "append 고수 추가")

mzfood.insert(0, "탕후루")
print(mzfood, "insert 0번에 탕후루 추가")

# 값 지우기
del mzfood[-1]
print(mzfood, "del 고수 제거")

del mzfood[0]
print(mzfood, "del 탕후루 제거")

# 값 바꾸기
mzfood[-2] = "양고기"
print(mzfood, "-2번 양고기로 바꿈")

# 문자열 시퀀스 아니냐, 그럼 저거 다 될 듯
인사 = "반가워"
인사[2] = "웡" # 안됨.
```

```python
# 왜 reverse, sort는 print문을 쓰면 none이 나올까?
print(mzfood)

mzfood.reverse() # 역순으로 변경
print(mzfood)
print(mzfood.reverse())

mzfood.sort() # 오름차순으로 정렬
print(mzfood)
print(mzfood.sort())
```

```python
# 2차원 리스트에 대해서도 공부 필수!!!

양파같은 = [[2,0], [3,1]]
print(양파같은)

# 1차원
print(양파같은[0])
print(양파같은[1])

# 2차원
print(양파같은[0][0])
print(양파같은[0][1])
print(양파같은[1][0])
print(양파같은[1][1])

# 추가
양파같은.append([4,1])
양파같은.append([7,1])
print(양파같은)

# 삭제
del 양파같은[2]
print(양파같은)
del 양파같은[1][0]
print(양파같은)
```
