## 시퀀스 자료형
* 문자열 :
    * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등 ...)
    * 슬라이싱( (:), (::) )
    * 요소들로 이루어져 가능했던 기능들(요소 합치기, 요소 찾기 등 ... )
* list :
    * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등 ...)
    * 슬라이싱( (:), (::) )
    * 요소들로 이루어져 가능했던 기능들(요소 합치기, 요소 찾기 등 ... )
    * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())
* tuple :
    * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
    * 값을 추가, 삭제, 변경할 수 없기 때문에 사용하지 못하는 메소드가 많다.
    * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값, 값)
    * 튜플은 소괄호!
* dict(딕셔너리) :
    * list, tuple, str, range(범위 안에 연속된 데이터를 만드는 시퀀스)  
    -> 데이터의 연관성이 없음
    * dict는 여러개의 값이 일렬로 정렬 되면서도 값 끼리의 연관성이 존재  
    (예 : 게임 캐릭터 힘-30, 지능-20, 체력-40, 민첩-90)
    * 생성 방법은 중괄호

```python
# list [] , tuple () , dict {}

짱짱맨 = {
    "힘" : 30,
    "지능" : 20,
    "체력" : 40,
    "민첩" : 90
}
print(짱짱맨)

# {키:값}
# {key1:value1, key2:value2}

# 시퀀스 데이터는 인덱싱을 이용해 데이터를 찾았습니다.
# 딕셔너리는 숫자가 아닌 키값을 이용해 데이터를 찾습니다.
짱짱맨["힘"]
print(f'짱짱맨님은 힘이 {짱짱맨["힘"]}입니다.')
```

```python
# 딕셔너리를 만드는 다양한 방법

# 빈 딕셔너리 만들기
딕셔너리 = {}
print(딕셔너리)

딕셔너리2 = dict()
print(딕셔너리2)

# 값을 넣어서 만드는 방법
# kict() 생성시 값 넣는 기호 "", : 주의해주세요
딕셔너리3 = dict(힘=30, 지능=20, 체력=40, 민첩=90)
print(딕셔너리3)

# zip() : 병렬 처리를 도와준다.
딕셔너리4 = dict(zip(["힘", "지능", "체력", "민첩"], [30, 20, 40, 90]))
print(딕셔너리4)
```

```python
# 값의 중복은 O, 키의 중복은 X

짱짱맨 = {
    "힘" : 30, # 씹힘
    "지능" : 20,
    "체력" : 40,
    "민첩" : 90,
    "지구력" : 90,
    "힘" : 50 # 제일 뒤의 키값으로 벨류 정해짐.
}
print(짱짱맨)

"체력" in 짱짱맨
```

```python
# 딕셔너리의 다양한 메소드

print(짱짱맨.keys())
print(짱짱맨.values())

print(짱짱맨.items())
print(type(짱짱맨.items()))


리스트화된_짱짱맨 = list(짱짱맨.items())
print(리스트화된_짱짱맨)
print(리스트화된_짱짱맨[3])
print(리스트화된_짱짱맨[3][1])
```
