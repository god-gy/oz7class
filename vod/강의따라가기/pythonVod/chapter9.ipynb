```python
# 형봔환 - int(), float(), str(), bool()

정수 = input()
정수 = int(정수)
print(정수, type(정수))

실수 = input()
실수 = float(실수)
print(실수, type(실수))
```

```python
정수 = int(input())
print(정수, type(정수))

실수 = float(input())
print(실수, type(실수))
```

```python
넌센스, 정답 = input().split(",")
print(f'넌센스 퀴즈 : {넌센스}')
print(f'정답 퀴즈 : {정답}')
```

```python
# 시퀀스 자료형에 다 지원되는 메소드
번호1, 번호2, 번호3, 번호4 = input().split(",")

번호1 = int(번호1)
번호2 = int(번호2)
번호3 = int(번호3)
번호4 = int(번호4)

print(번호1, 번호2, 번호3, 번호4)
print(type(번호1), 번호2, 번호3, 번호4)
print(f'번호1 : {번호1}')
print(f'번호2 : {번호2}')
```

```python
# map()

번호1, 번호2, 번호3, 번호4 = map(int, input().split(","))

print(type(번호1), 번호2, 번호3, 번호4)
```

```python
# 레거시 코드 : 누군가 남겨두고간 코드
# 리팩토링 : 결과의 변경 없이 코드의 구조를 재조정 하는 것.

fish_p, fish_s, fish_j = input('붕어빵(팥), 붕어빵(슈크림), 붕어빵(잡채)를 순서대로 개수를 입력하고 꼭 쉼표를 입력해라. 안사는 붕어빵에는 0을 넣어줘.').split(',')
fish_p_sell, fish_s_sell, fish_j_sell = 2000, 2500, 3000
order_s, order_j = 2, 3  # 리팩토링 해보세요
order_s_pay, order_j_pay = fish_s_sell * 2, fish_j_sell * 3
total = order_s + order_j
total_pay = order_s_pay + order_j_pay

print(f'주문이 완료되었습니다. {fish_s}{order_s}개, {fish_j}{order_j}개 총 {total}개 결제 금액은 {total_pay}원 입니다.')
```
