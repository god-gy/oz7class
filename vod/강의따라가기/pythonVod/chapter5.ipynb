## 변수란? ( 기본 중에 기본, 근본 중에 근본, 코어 중에 코어)

-> 하나의 변수에 하나의 값을 넣을때 : 변수이름 = 값

-> 두개의 변수에 하나의 값을 각각 넣을때 : 변수이름1, 변수이름2 = 값1, 값2

```python
# 하나의 값을 변수에 넣을때
num = 1
#print(num)

# 두개의 값을 변수에 넣을 때
num1, num2 = 1, 2
print(num1)
print(num2)

print(num1, num2)
```

### 변수에 값을 넣는 것 => 할당

sum = 1+2 => 1+2의 계산된 결과를 sum이라는 변수에 할당한다.  
sum은 변수명이고, "=" 같다는 의미가 아니라 할당(assignment)이라고 한다.  
"=" : 할당 연산자  
"+" : 산술 연산자  
+= 산술 후 바로 할당하는 문법  
/=, -=, *= 모두 가능.

```python
# 변수를 사용하면서 사용 가능한 문법

sum = 1+2
sum = sum+3
sum = sum+4
sum = sum+5

print(sum)

sum2 = 1+2
sum2 += 3
sum2 += 4
sum2 += 5

print(sum2)
```
