```
if 파이썬을 정복한다면:
    개발자로 취직할 수 있다.
```




###2. if문 문법과 주의사항 (콜론:, 들여쓰기)
* 파이썬에서 들여쓰기는 문법 그 자체


```
question = input("노루가 다니는 길은?")

if question == "노루웨이":
    print("정답입니다!")
else:
    print("오답입니다.")
```



````python
#2. if문 문법과 주의사항 (콜론:, 들여쓰기)
개수 = int(input("붕어빵 몇개 사실껀가요?"))

if 개수 > 3:
    print("개당 1800원 입니다.")
````

````python
#3. if문의 흐름 (작동 원리)
# 코드는 위에서 아래로 작동합니다.

# 2)개수라는 변수에 담음  1) 입력을 받음
개수 = int(input("붕어빵 몇개 사실껀가요?"))

# 3) 개수 변수의 값이 3보다 큰지 비교 (4>3) True or False
if 개수 > 3:
    print("개당 1800원 입니다.") # 4) 조건이 True면 if문 안의 코드가 실행됨.


if 개수 <= 3:
    print("개당 2000원 입니다.")
````

````python
#4. if 중첩문

의문의_숫자 = 10

if 의문의_숫자 > 5:
    if 의문의_숫자 < 15:
        print("의문의 숫자는 5보다 크고 10보다 작습니다.")
````

````python
#5. if ~else 문법과 주의사항 (콜론, 들여쓰기)
# True 일때는 if문 안에 작성된 코드가 실행됨.
# False 일때는 else문 안에 작성된 코드가 실행됨.

개수 = int(input("붕어빵 몇개 사실껀가요?"))

if 개수 > 3:
    print("개당 1800원 입니다.")
else:
    print("개당 2000원 입니다.")
````

````python
#6. if ~else문 흐름 (작동 원리)

개수 = int(input("붕어빵 몇개 사실껀가요?"))

if 개수 > 3:
    print("개당 1800원 입니다.")
else:
    print("개당 2000원 입니다.")
````

````python
#7. 조건문 조건 (True, False)

# True
if True:
    print(True)
else:
    print(False)

# False
if False:
    print(True)
else:
    print(False)

# None
if None:
    print(True)
else:
    print(False)

# 0
if 0:
    print(True)
else:
    print(False)

# 1
if 1:
    print(True)
else:
    print(False)

# 1.1
if 1.1:
    print(True)
else:
    print(False)

# 'hello'
if 'hello':
    print(True)
else:
    print(False)

# ''
if '':
    print(True)
else:
    print(False)

# not True
if not True:
    print(True)
else:
    print(False)

# not ''
if not '':
    print(True)
else:
    print(False)
````

````python
#8. 조건식 여러개 설정하기

a, b = 10, 30

if a ==10 and b ==30:
    print("동일합니다.")
else:
    print("동일하지 않습니다.")

if a ==10 or b ==30:
    print("동일합니다.")
else:
    print("동일하지 않습니다.")

if a >=10 and a <20:
    print("10보다 크고 20보다 작습니다.")

if 10<=a <20:
    print("10보다 크고 20보다 작습니다.")
````

````python
#9. elif~if문

맛 = input("붕어빵 맛을 골라주세요(팥, 슈크림, 잡채) : ")

if 맛 == "팥":
    print(f'{맛}을 선택하셨습니다.')
elif 맛 == "슈크림":
    print(f'{맛}을 선택하셨습니다.')
elif 맛 == "잡채":
    print(f'{맛}을 선택하셨습니다.')
else:
    print("존재하지 않는 맛입니다.")
````

````python
#10. 예제로 if문 익혀보기
# 문제 : 단어 a의 가운데 글자를 반환하는 코드
# 조건 : 단어의 길이가 짝수라면, 가운데 두글자를 출력하면 됩니다.

# 입력값 1 : "abcde" => 출력값 : c
# 입력값 2 : "qwer" => 출력값 : we

a = input()

# 코드를 작성해주세요.

가운데_인덱스 = int(len(a)/2)

if len(a) % 2 != 0: # 홀수일 때
    print(a[가운데_인덱스])
else:
    print(a[가운데_인덱스-1 : 가운데_인덱스+1])
````
