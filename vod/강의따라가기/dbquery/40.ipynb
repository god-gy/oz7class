```python
# 데이터베이스 선택 (없으면 자동 생성)
db = client['example_db']
db
```

```python
# 콜렉션 선택 (없으면 자동 생성)
collection = db['example_collection']
collection
```

### CRUD

```python
# 새 문서 삽입
example_document = {"name": "Amy", "age": 50, "city": "Callifornia"}

collection.insert_one(example_document)
```

```python
# 모든 문서 조회
for doc in collection.find():
    print(doc)
```

```python
# 조건에 맞는 문서 조회
query = {"name": "John"}
for doc in collection.find(query):
    print(doc)
```

```python
# 하나의 문서 업데이트
collection.update_one({"name": "John"}, {"$set": {"age": 31}})

for doc in collection.find():
    print(doc)
```

```python
# 여러 문서 업데이트
collection.update_many({"name": "John"}, {"$set": {"age": 32}})

for doc in collection.find():
    print(doc)
```

```python
# 하나의 문서 삭제
collection.delete_one({"name": "John"})

for doc in collection.find():
    print(doc)
```

```python

# 여러 문서 삭제
collection.delete_many({"name": "John"})

for doc in collection.find():
    print(doc)
```

```python
# 콜렉션 삭제
db.drop_collection("example_collection")
```

```python
# 데이터베이스 삭제
client.drop_database("example_db")
```
