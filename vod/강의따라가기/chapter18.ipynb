````python
# 클래스 구조

class 클래스이름:
    def 매서드(self):
        코드
````

````python
class Person:
    def hello(self):
        print('수강생 여러분')

호출 = Person() # 할당된 호출 : 인스턴스
호출.hello() # .index(), .count()
````

### self가 무엇이냐

인스턴스 자기 자신을 의미한다.  
여기서 __init__의 매개변수 self에 들어가는 값은  
Person() 이라고 할 수 있음


```
class Person:
    def __init__(self):     # self -> james = Person()
        self.hello = "안녕하세요"

    def greeting(self):     # self -> james.greeting()
        print(self.hello)
```



````python
class 클래스이름:
    def __init__(self, 매개변수1, 매개변수2):
        self.속성1 = 매개변수1
        self.속성2 = 매개변수2
````

````python
class Person:
    def __init__(self, name, age, address):
        self.hello = '안녕하세요.'
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print('{0} 저는 {1}입니다.'.format(self.hello, self.name))

maria = Person('마리아', 20, '서울시 서초구 반포동')
maria.greeting()    # 안녕하세요. 저는 마리아입니다.

print('이름:', maria.name)       # 마리아
print('나이:', maria.age)        # 20
print('주소:', maria.address)    # 서울시 서초구 반포동
````

## 클래스의 상속

기반클래스(base class) -> 파생 클래스(derived class)

````python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print("멈췄습니다.")

class HybridCar(Car):       # 매개변수로 Car class를 상속 받음
    battery = 1000
    batteryKM = 300


k5 = HybridCar()
print(k5.maxspeed)

print(dir(k5))      # 'maxpeople', 'maxspeed', 'move', 'stop'
````

````python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print("멈췄습니다.")

class HybridCar(Car):       # Car 상속
    battery = 1000
    batteryKM = 300

class ElectricCar(HybridCar):      # Car를 상속받은 HybridCar를 상속
    battery = 1000
    batteryKM = 300

k5 = HybridCar()
print(k5.maxspeed)

electricCark5 = ElectricCar()
print(electricCark5.maxspeed)

print(dir(k5))
````

````python
# 메서드 오버라이딩 -> 가장 마지막 메서드만 인정됨

class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):      # 1
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print("멈췄습니다.")

class HybridCar(Car):
    battery = 1000
    batteryKM = 300

class ElectricCar( HybridCar):
    battery = 1000
    batteryKM = 300
    def move(self, x):      # 2 -> 이게 실행됨
        print(self.batteryKM, '만큼 달릴 수 있습니다.')
        print(x, '스피드로 달리고 있습니다.')

k5 = HybridCar()
k5.maxspeed

electricCark5 = ElectricCar()
electricCark5.maxspeed
electricCark5.battery
electricCark5.move(10)      # move 메소드 실행

# print(dir(k5))
````
