# AWS (Amazon Web Services) 개요

---

**AWS(Amazon Web Services)**는 [Amazon.com](http://Amazon.com)의 클라우드 컴퓨팅 사업부로, 전 세계에서 가장 포괄적이고 널리 사용되는 클라우드 플랫폼입니다.

## 핵심 특징

- **서비스 제공 방식**: 인터넷을 통해 다양한 IT 리소스(컴퓨팅 파워, 스토리지, 데이터베이스, 머신러닝 등)를 온디맨드로 제공
- **요금 체계**: 사용한 만큼만 비용을 지불하는 종량제(Pay-as-you-go) 모델
- **고객 규모**: 스타트업부터 대기업, 정부 기관까지 수백만 고객 보유
- **서비스 수**: 200개 이상의 서비스 제공

## AWS의 주요 이점

**확장성 및 유연성**

필요에 따라 컴퓨팅 용량과 스토리지를 즉시 조절 가능

**비용 효율성**

초기 인프라 투자 없이 사용한 리소스에 대해서만 비용 지불

**안정성 및 보안**

전 세계에 분산된 데이터 센터와 강력한 보안 기능 제공

**글로벌 인프라**

여러 리전(Region)과 가용 영역(Availability Zones)을 통해 전 세계에 신속한 배포 가능

# AWS 주요 서비스

## 컴퓨팅 (Compute)

### Amazon EC2 (Elastic Compute Cloud)

- 클라우드에서 제공하는 가상 서버 인스턴스
- 사용자가 원하는 시점에 컴퓨팅 성능(CPU, 메모리, 스토리지, 네트워크)을 유연하게 조절 가능
- 물리적 서버 없이 애플리케이션 실행 환경 제공

### AWS Lambda

- 서버리스(Serverless) 컴퓨팅 서비스
- 서버 관리 없이 코드 실행 가능
- 이벤트 기반으로 자동 실행 및 확장

## 스토리지 (Storage)

### Amazon S3 (Simple Storage Service)

- 무제한 용량의 확장 가능한 객체 스토리지 서비스
- 파일, 이미지, 동영상 등 모든 유형의 데이터 저장
- 높은 내구성(99.999999999%)과 가용성 제공

## 데이터베이스 (Database)

### Amazon RDS (Relational Database Service)

- 관계형 데이터베이스 관리 서비스
- 지원 엔진: MySQL, PostgreSQL, MariaDB, Oracle, SQL Server, Amazon Aurora
- 자동 백업, 패치, 복제 기능 제공

### Amazon DynamoDB

- 완전 관리형 NoSQL 데이터베이스 서비스
- 빠른 성능과 유연한 스키마
- 자동 확장 기능

## 네트워킹 및 콘텐츠 전송

### Amazon CloudFront

- 콘텐츠 전송 네트워크(CDN) 서비스
- 전 세계 엣지 로케이션에서 콘텐츠를 캐싱하여 사용자에게 빠르게 전송
- 지연 시간 최소화 및 전송 속도 향상

### Amazon VPC (Virtual Private Cloud)

- AWS 클라우드 내에서 논리적으로 격리된 가상 네트워크
- 사용자가 정의한 네트워크 환경에서 AWS 리소스 실행
- 서브넷, 라우팅 테이블, 게이트웨이 등을 직접 제어

## 보안 및 액세스 관리

### AWS IAM (Identity and Access Management)

- AWS 리소스에 대한 액세스를 안전하게 관리하는 핵심 서비스
- 사용자 권한과 보안 자격 증명을 중앙에서 통합 관리
- 최소 권한 원칙(Principle of Least Privilege) 구현

# 클라우드 아키텍처 (Cloud Architecture)

## 아키텍처란?

**아키텍처(Architecture)**는 시스템을 구성하는 요소들과 그 관계, 설계 원칙을 의미합니다. IT 분야에서는 시스템의 청사진 또는 설계도에 해당합니다.

## 클라우드 아키텍처의 목적

AWS, Azure, GCP와 같은 클라우드 플랫폼에서 애플리케이션과 인프라를 효율적으로 구성하는 방법을 계획합니다.

**주요 목표**

- 확장성(Scalability) 확보
- 고가용성(High Availability) 보장
- 비용 효율성(Cost Efficiency) 최적화
- 보안성(Security) 강화

## 클라우드 아키텍처 설계 시 고려 사항

1. **로드 밸런싱**: 트래픽을 여러 서버에 분산하여 처리
2. **네트워크 구성**: VPC, 서브넷, 보안 그룹 설계
3. **데이터베이스 배치**: 읽기 전용 복제본, 다중 AZ 배포 등
4. **서버리스 기술 도입**: Lambda, API Gateway 활용
5. **모니터링 및 로깅**: CloudWatch, CloudTrail 설정

## 아키텍처의 핵심 구성 요소

**구조(Structure)**

시스템의 구성 요소와 배치 방식

**원칙과 제약 조건(Principles & Constraints)**

설계를 이끄는 기본 규칙과 제한 사항

예: "모든 서비스는 독립적으로 배포 가능해야 한다"

**품질 속성(Quality Attributes)**

시스템이 달성해야 하는 목표

예: 성능, 보안성, 확장성, 안정성

## 주요 아키텍처 패턴

### 마이크로서비스 아키텍처(MSA)

- 애플리케이션을 작은 독립적인 서비스로 분리
- 각 서비스는 독립적으로 개발, 배포, 확장 가능

### 모놀리식 아키텍처

- 모든 기능이 하나의 애플리케이션으로 통합
- 단순하지만 확장성과 유지보수에 제약

### 이벤트 기반 아키텍처

- 이벤트 발생에 따라 시스템이 반응
- 느슨한 결합(Loose Coupling)으로 유연성 향상

# AWS IAM (Identity and Access Management)

## IAM 개요

**AWS IAM**은 AWS 리소스에 대한 접근을 안전하게 관리하는 핵심 서비스입니다.

**주요 기능**

- **누가(Who)**: 사용자, 애플리케이션, AWS 서비스
- **어떤 리소스에(What)**: EC2, S3, DynamoDB 등
- **어떤 작업을(Action)**: 읽기, 쓰기, 삭제 등
- **접근 권한을 제어**

## IAM의 네 가지 핵심 구성 요소

### 1. 사용자(Users)

- AWS 서비스와 상호작용하는 개별 사람 또는 애플리케이션
- 각 사용자는 고유한 자격 증명을 보유
- **자격 증명 유형**:
    - AWS 콘솔 접속용 비밀번호
    - 프로그래밍 방식 접근용 액세스 키(Access Key ID + Secret Access Key)

### 2. 그룹(Groups)

- 여러 사용자를 하나의 단위로 묶어 관리하는 컨테이너
- 그룹에 정책을 연결하면 소속된 모든 사용자가 동일한 권한을 자동으로 상속
- **예시**: 개발자 그룹(Developers), 관리자 그룹(Admins), QA팀 그룹

### 3. 역할(Roles)

- AWS 서비스 또는 외부 사용자에게 **임시 권한**을 부여하는 자격 증명
- 영구적인 자격 증명(비밀번호, 액세스 키)을 갖지 않음
- 필요할 때만 임시로 권한을 부여받아 작업 수행 후 반납
- **주요 사용 사례**:
    - EC2 인스턴스가 S3 버킷에 접근
    - Lambda 함수가 DynamoDB 테이블을 읽고 쓰기
    - 다른 AWS 계정의 사용자에게 임시 접근 권한 부여

### 4. 정책(Policies)

- AWS 리소스에 대한 접근 권한을 **JSON 형식**으로 정의한 문서
- **허용(Allow)** 또는 **거부(Deny)** 규칙 지정
- 정책은 사용자, 그룹, 역할에 연결(Attach)되어 적용

**정책 예시 구조**:

```json
{
  "Version": "2012-10-17",
  "Statement": [
    {
      "Effect": "Allow",
      "Action": "s3:GetObject",
      "Resource": "arn:aws:s3:::my-bucket/*"
    }
  ]
}
```

## IAM 권한 부여 방법

### 사용자에게 권한 부여

**User ← Policy 연결**

- 특정 사용자에게 직접 정책을 연결하여 권한 부여
- 소규모 환경에 적합

### 그룹을 통한 권한 부여

**User → Group ← Policy 연결**

- 그룹에 정책을 연결하고 사용자를 그룹에 추가
- 대규모 환경에서 권한 관리 효율화

### AWS 서비스에 권한 부여

**Service → Role ← Policy 연결**

- AWS 서비스(EC2, Lambda 등)가 다른 서비스를 사용할 수 있도록 역할에 정책 연결
- 예: EC2 인스턴스가 S3에 접근하려면 EC2 인스턴스에 S3 접근 권한이 있는 역할 부여

## IAM 보안 모범 사례

### 최소 권한 원칙(Principle of Least Privilege)

- 사용자나 서비스가 작업을 완료하는 데 **필요한 최소한의 권한만** 부여
- 불필요한 권한은 보안 위험 증가

### 루트 사용자 보호

- 루트 사용자는 모든 권한을 가지므로 일상적인 작업에 사용 금지
- MFA(Multi-Factor Authentication) 활성화 필수

### 정기적인 권한 검토

- 사용하지 않는 사용자, 액세스 키, 정책을 주기적으로 제거

### MFA 활성화

- 중요한 작업을 수행하는 사용자는 반드시 MFA 설정

# Amazon S3 (Simple Storage Service)

## S3 개요

**Amazon S3**는 AWS의 대표적인 객체 스토리지 서비스로, 무제한 용량의 데이터를 저장하고 검색할 수 있습니다.

**핵심 특징**

- **객체(Object) 단위 저장**: 파일을 객체로 관리
- **사실상 무제한 용량**: 확장 가능한 스토리지
- **높은 내구성**: 99.999999999%(11 nines) 데이터 내구성 보장
- **고가용성**: 여러 가용 영역에 자동 복제
- **다양한 접근 제어**: 버킷 정책, IAM 정책, ACL 등

## S3의 주요 구성 요소

### 버킷(Bucket)

- S3에 저장되는 객체의 최상위 컨테이너
- **버킷 이름은 전 세계에서 고유**해야 함 (리전과 무관하게 중복 불가)
- 버킷은 특정 리전에 생성되지만, 이름은 글로벌 네임스페이스 사용

### 객체(Object)

- S3에 저장되는 개별 파일
- 각 객체는 **키(Key)**, **값(Value)**, **메타데이터**로 구성
- 최대 객체 크기: 5TB

## S3 권한 관리

S3에서 가장 중요한 부분은 **읽기/쓰기 권한 관리**입니다.

### 1. 퍼블릭 액세스 차단 설정

- **활성화**: 외부에서 접근 불가 (완전 비공개)
- **비활성화**: 버킷 정책에 따라 외부 접근 허용 가능

### 2. 버킷 정책(Bucket Policy)

- JSON 형식으로 버킷 및 객체에 대한 접근 권한 정의
- 특정 사용자, IP 주소, AWS 서비스에 대한 접근 제어

**버킷 정책 예시**:

```json
{
  "Version": "2012-10-17",
  "Statement": [
    {
      "Sid": "PublicReadGetObject",
      "Effect": "Allow",
      "Principal": "*",
      "Action": "s3:GetObject",
      "Resource": "arn:aws:s3:::be-15-bucket/*"
    }
  ]
}
```

**정책 구성 요소 설명**

- **Version**: 정책 언어 버전 (항상 "2012-10-17" 사용)
- **Statement**: 권한 규칙의 배열
- **Sid (Statement ID)**: 각 Statement를 식별하는 ID (선택 사항)
- **Effect**: "Allow" (허용) 또는 "Deny" (거부)
- **Principal**: 권한을 부여받는 주체 ("*"는 모든 사용자)
- **Action**: 허용/거부할 작업 (예: s3:GetObject, s3:PutObject)
- **Resource**: 권한이 적용되는 리소스 ARN
    - `arn:aws:s3:::be-15-bucket/*`: 버킷 내 모든 객체
    - `arn:aws:s3:::be-15-bucket`: 버킷 자체

## S3 접근 제어 시나리오

### 완전 공개로 설정하기

1. **퍼블릭 액세스 차단**: 비활성화
2. **버킷 정책 추가**: Principal을 "*"로 설정하고 s3:GetObject 허용

**사용 사례**: 정적 웹사이트 호스팅, 공개 이미지/파일 배포

### 완전 비공개로 설정하기

1. **퍼블릭 액세스 차단**: 활성화
2. **버킷 정책 삭제** 또는 특정 IAM 사용자/역할에만 접근 허용

**사용 사례**: 민감한 데이터, 내부 백업 파일, 로그 저장

## S3 사용 사례

### 정적 웹사이트 호스팅

- HTML, CSS, JavaScript 파일을 S3에 업로드
- S3 정적 웹사이트 호스팅 기능 활성화
- CloudFront와 결합하여 전 세계에 빠르게 배포

**정적 사이트 vs 동적 사이트**

- **정적 사이트**: 내용이 고정되어 있으며 서버 측 처리 없음 (예: 블로그, 포트폴리오)
- **동적 사이트**: 사용자 입력이나 데이터베이스에 따라 내용이 변경됨 (예: 로그인 시스템, SNS)

### 데이터 백업 및 아카이빙

- 자동 백업 스크립트와 연동
- S3 Glacier로 장기 보관 시 비용 절감

### 애플리케이션 데이터 저장

- 사용자 업로드 파일, 이미지, 동영상 저장
- CDN과 연계하여 빠른 콘텐츠 전송

# Amazon EC2 (Elastic Compute Cloud)

## EC2 개요

**Amazon EC2**는 AWS에서 제공하는 가상 서버 서비스로, 클라우드 환경에서 컴퓨팅 리소스를 유연하게 사용할 수 있게 합니다.

**핵심 특징**

- **가상 서버 인스턴스 제공**: 물리적 서버 없이 서버 환경 구성
- **유연한 확장성**: CPU, 메모리, 스토리지, 네트워크 용량을 필요에 따라 조절
- **다양한 인스턴스 타입**: 범용, 컴퓨팅 최적화, 메모리 최적화, GPU 인스턴스 등
- **종량제 요금**: 사용한 시간만큼만 비용 지불

## EC2 네트워크 구조

모든 EC2 인스턴스는 반드시 **VPC(Virtual Private Cloud)** 내에 배치되어야 합니다.

**구조 흐름**

```
VPC (가상 네트워크)
  ↓
보안 그룹 (Security Group - 방화벽)
  ↓
EC2 인스턴스
```

### VPC (Virtual Private Cloud)

- AWS 클라우드 내에서 논리적으로 격리된 가상 네트워크
- 사용자가 IP 주소 범위, 서브넷, 라우팅 테이블, 게이트웨이를 직접 정의

### 보안 그룹 (Security Group)

- EC2 인스턴스에 대한 가상 방화벽
- **인바운드 규칙**: 외부에서 인스턴스로 들어오는 트래픽 제어
- **아웃바운드 규칙**: 인스턴스에서 외부로 나가는 트래픽 제어

**보안 그룹 설정 예시**

- SSH (포트 22): 특정 IP에서만 접근 허용
- HTTP (포트 80): 모든 IP에서 접근 허용
- HTTPS (포트 443): 모든 IP에서 접근 허용

## EC2 인스턴스 생성 시 주요 설정

### 1. 키 페어 (Key Pair) 생성

**키 페어란?**

- EC2 인스턴스에 안전하게 접속하기 위한 암호화 키
- **프라이빗 키(.pem 또는 .ppk)**: 사용자가 보관
- **퍼블릭 키**: AWS가 인스턴스에 저장

**키 페어 유형**

- **RSA**: 가장 널리 사용되는 암호화 알고리즘
- **ED25519**: 최신 암호화 알고리즘, 더 짧은 키로 높은 보안성 제공

**프라이빗 키 파일 형식**

- **.pem**: macOS, Linux에서 사용
- **.ppk**: Windows (PuTTY)에서 사용

### 2. 사용자 데이터 (User Data)

**사용자 데이터란?**

- EC2 인스턴스가 처음 시작될 때 자동으로 실행되는 스크립트
- 초기 설정, 소프트웨어 설치, 서비스 시작 등을 자동화

**사용자 데이터 예시 (웹 서버 자동 설치)**:

```bash
#!/bin/bash
yum install httpd -y
systemctl start httpd
systemctl enable httpd
echo "<h1>Hello, UserData</h1>" > /var/www/html/index.html
```

**스크립트 설명**

- `#!/bin/bash`: Bash 셸 스크립트 지정
- `yum install httpd -y`: Apache HTTP 서버 설치 (자동 승인)
- `systemctl start httpd`: 웹 서버 시작
- `systemctl enable httpd`: 재부팅 시 자동으로 웹 서버 시작되도록 설정
- `echo "<h1>Hello, UserData</h1>" > /var/www/html/index.html`: 기본 HTML 페이지 생성

### 3. S3와 연동하는 사용자 데이터

S3 버킷에 저장된 파일을 EC2 인스턴스로 복사하여 웹 서버에 배포하는 예시:

```bash
#!/bin/bash
yum install httpd php -y
aws s3 cp s3://oz-be15-bucket-010/index.php /var/www/html --region ap-northeast-2
systemctl start httpd
systemctl enable httpd
```

**주의사항**

- EC2 인스턴스에 S3 접근 권한이 있는 **IAM 역할**이 연결되어 있어야 함
- S3 버킷 정책 또는 IAM 정책에서 `s3:GetObject` 권한 필요

# Auto Scaling Group (ASG)

## 📌 ASG란?

Auto Scaling Group은 서버(EC2)의 개수를 **자동으로 늘리거나(Scale-out)**

**자동으로 줄이는(Scale-in)** AWS의 오토스케일링 기능.

## 📌 왜 필요한가?

- 트래픽이 갑자기 증가해도 서버 다운 없이 처리
- 트래픽이 적을 때는 인스턴스를 줄여 비용 절약
- 인스턴스가 죽어도 자동으로 다시 생성 → 높은 가용성

---

## ✔ ASG 동작 예시 시나리오

### 🧩 시나리오 1 — 출퇴근 시간대 트래픽 증가

- 평소에는 EC2 2대가 서비스 운영
- 오전 8시, 오후 6시에 사용자 증가 → CPU 70% 초과
- **ASG가 자동으로 인스턴스 3대를 추가 생성**
- 트래픽 감소 후 CPU 30% 이하 →

    **인스턴스 3대를 정리하고 다시 2대로 감소**


### 🧩 시나리오 2 — 서버 장애 발생

- EC2 인스턴스 하나가 다운됨(Health Check 실패)
- ALB가 해당 서버로 트래픽 전달 중단
- ASG가 **새로운 EC2 인스턴스를 자동 생성**
- 서비스 장애 없이 복구됨

---

## ✔ ASG 구성 요소

### 1. Launch Template

어떤 환경의 EC2를 만들지 지정

### 2. Desired Capacity (원하는 인스턴스 수)

초기 실행할 인스턴스 개수 예: 2대

### 3. Min / Max Capacity

- Min: 최소 2대
- Max: 최대 5대

### 4. Scaling Policy

- CPU 60% 이상이면 1대 추가
- CPU 20% 이하이면 1대 감소

---

## 📘 ASG 생성 예시 설정

```yaml
ASG 이름: web-asg
Availability Zone: ap-northeast-2a, 2c
Desired Capacity: 2
Min Capacity: 2
Max Capacity: 5
Load Balancer: ALB (web-alb)
Health Check Type: ELB Health Checks
Scaling Policy:
   - CPU > 60% → +1 Instance
   - CPU < 20% → -1 Instance
Launch Template: web-template-v1

```

---

## 오토스케일링의 고가용성 (High Availability)

## 📌 고가용성이란?

서비스가 **고장나도 멈추지 않는 상태를 유지하는 능력**을 의미한다.

ASG + Multi-AZ 구성은 AWS에서 고가용성을 구현할 때 가장 핵심적인 구조.

### ✔ 왜 ASG가 고가용성을 보장할까?

1. **인스턴스 장애 시 자동 복구**
    - 특정 EC2가 다운되거나 Health Check에 실패하면

        → ASG가 자동으로 새 인스턴스 생성

2. **Availability Zone(AZ) 장애가 발생해도 운영 가능**
    - ap-northeast-2a가 다운되면

        → ASG가 2c, 2d 같은 다른 AZ에 인스턴스 재배치

3. **ALB와 함께 사용할 때 더 강력**
    - ALB는 정상 인스턴스로만 트래픽을 보내고
    - ASG는 죽은 인스턴스를 복구

즉,

**“죽으면 다시 만들고, 장애가 난 곳은 자동으로 피해서 실행하는 구조” → 고가용성 보장**

---

## 📘 고가용성 구성 예시

```yaml
ASG: web-asg
AZ 선택: ap-northeast-2a, 2c, 2d
Desired: 2
Min: 2
Max: 6
ALB: web-alb
Health Check: ELB 기반

```

### 상황 예시

- 2a AZ 장애 발생 → 2a 인스턴스 1개 죽음
- ASG가 자동으로 2c 또는 2d에 새로운 인스턴스 1개 생성
- ALB는 정상 인스턴스만 사용하여 서비스 중단 없음

# 7️⃣ Auto Scaling — 가용영역(AZ) 및 서브넷 선택

## 📌 ASG에서 가용영역을 왜 여러 개 선택해야 하나?

ASG는 EC2 인스턴스를 **여러 AZ에 분산 배치**하여

특정 AZ에서 장애가 나도 서비스가 유지되도록 한다.

만약 1개 AZ만 선택하면?

❌ 그 AZ에 문제가 생기면 서비스 전체가 중단됨 (싱글 포인트 고장)

---

## 📌 ASG 서브넷 선택 시 주의사항

### ✔ 1. 반드시 **Public Subnet** 또는 **Private Subnet** 중 목적에 맞게 선택

- ALB 뒤에 있는 웹 서버라면 → **Private Subnet**
- Bastion Host 또는 Public Web Server라면 → **Public Subnet**

### ✔ 2. 최소 2개 이상의 AZ 선택

예:

- ap-northeast-2a
- ap-northeast-2c

AWS 권장은 2~3개 AZ

---

## 📘 서브넷 설정 예시 (웹 서버용)

```vbnet
ASG Subnets:
  - private-subnet-2a
  - private-subnet-2c
  - private-subnet-2d

설명:
  ALB는 Public Subnet 2개에 존재
  웹서버 ASG는 Private Subnet 3개에 분산 배치

```

### 실제 배치 흐름

- ALB(Public Subnet) → EC2(Private Subnet)
- ASG는 private-subnet-2a, 2c, 2d 중 헬스체크와 수량에 맞춰 분산 배치

# 8️⃣ Auto Scaling — 상태 확인(Health Check) & 유효시간(Grace Period)

---

## 📌 Health Check란?

ASG가 EC2 인스턴스를 감시하는 기능

ASG는 두 가지 health check를 사용할 수 있다:

### ✔ 1. EC2 Health Check

- OS-level 상태
- 시스템이 응답하는지
- 커널 패닉 여부

    등을 확인


### ✔ 2. ELB Health Check (권장)

- ALB Target Group에서 제공하는 헬스 체크
- HTTP 경로(/health) 응답 여부
- 포트 응답 정상 여부
- 트래픽을 처리 가능한 상태인지 확인

    → 실제 서비스 상태를 더 잘 반영함


---

## ✔ 둘의 차이

| 종류 | 특징 | 권장도 |
| --- | --- | --- |
| EC2 Health Check | 인스턴스 자체만 체크 | 보통 보조적으로 사용 |
| **ELB Health Check** | 서비스 상태까지 체크 (HTTP 응답) | **가장 권장되는 방식** |

---

## 📌 Health Check Grace Period (상태 확인 유예 시간)

### ✔ Grace Period(유예시간)란?

ASG가 **인스턴스를 생성한 뒤 즉시 health check를 하지는 않고**,

일정 시간(기본 300초) 기다렸다가 health check를 시작하는 기능.

### 왜 필요할까?

- User Data에서 앱 설치/도커 실행/서버 빌드 등 시간이 걸림
- 인스턴스가 완전히 준비되기 전에 Health Check를 하면

    → 정상인데도 오류로 판단해서 "계속 재생성 루프" 발생


그래서 ASG는 유예 시간을 두고,

그 동안은 health check 실패해도 무시한다.

---

## 📘 Grace Period 예시 설정

```yaml
Health Check Type: EC2 + ELB
Grace Period: 300 seconds (5 minutes)
Health Check Path: /health
Health Check Interval: 30 seconds
Healthy threshold: 2
Unhealthy threshold: 2

```

### 동작 흐름

1. 인스턴스 새로 생성
2. 300초 동안 health check 무시
3. 300초 이후 → 정상적으로 health check 시작
4. /health 응답 없으면 → 인스턴스를 교체

---

# 9️⃣ 상태 확인 시나리오 예시

### 🧩 예시 1 — 서버 초기 구동이 느린 경우

- User Data에서 Python 앱 설치, 의존성 설치 2~3분 소요
- Grace Period를 300초로 설정

    → 설치 중에는 오류로 판단하지 않음

    → 서버가 완전히 준비된 다음에만 health check 수행


### 🧩 예시 2 — 서버가 죽어서 헬스체크 실패

- 인스턴스가 /health에 200 응답 반환 안 함

    → ASG가 해당 인스턴스를 중지

    → 새로운 인스턴스 생성 후 Grace Period 동안 대기

    → 정상 확인 후 ALB에 연결

# Launch Template (시작 템플릿)

## 📌 Launch Template란?

ASG가 EC2를 생성할 때 사용되는 **설정 템플릿**.

즉, “이 템플릿대로 EC2를 만들어라”는 설계도.

---

## ✔ Launch Template에 포함되는 항목

- AMI (Amazon Linux 2 등)
- 인스턴스 유형 (t3.micro)
- Key Pair
- Security Group
- IAM Role
- Storage(EBS) 설정
- User Data (초기 실행 스크립트)

---

## 📘 Launch Template 예시 설정

```yaml
Template Name: web-template
Version: v1
AMI: Amazon Linux 2023 AMI
Instance Type: t3.micro
Security Group:
   - http, https inbound 허용
   - ssh는 bastion host만 허용
Key Pair: web-key
EBS Volume:
   - 20GB gp3
User Data:
   - Nginx 자동 설치
   - 서버 자동 부팅 스크립트

```

---

## 📘 User Data 예시 (웹 서버 자동 설치)

```bash
#!/bin/bash
yum install httpd -y

TOKEN=$(curl -X PUT "http://169.254.169.254/latest/api/token" \
    -H "X-aws-ec2-metadata-token-ttl-seconds: 21600")

INSTANCE_ID=$(curl -s -H "X-aws-ec2-metadata-token: $TOKEN" \
    http://169.254.169.254/latest/meta-data/instance-id)

echo "$INSTANCE_ID" > /var/www/html/index.html

systemctl enable --now httpd
```

# Elastic Load Balancer (ELB)

## 📌 ELB란?

엘라스틱 로드 밸런서.
들어오는 트래픽을 여러 인스턴스에 자동으로 분배하는 서비스.

- 장애 인스턴스는 자동 제외
- 헬스 체크 수행
- HTTPS 인증서 적용 가능(ACM 사용)

---

## ✔ ELB 종류와 특징

### ⭐ ALB (Application Load Balancer)

- Layer 7(HTTP/HTTPS)
- URL 기반 라우팅 가능

    /api → API 서버

    /admin → 관리자 서버


### ⭐ NLB (Network Load Balancer)

- Layer 4
- 초고속 처리
- WebSocket, TCP 기반 서비스에 적합

### ⭐ CLB (Classic Load Balancer)

- 구버전 → 거의 사용 X

---

## 🎯 ALB 헬스체크

ALB 헬스체크는 “지금 이 서버에 트래픽을 보내도 되는지”를 자동으로 판단해
장애 인스턴스를 제외하고 ASG에게 새로운 인스턴스를 만들도록 알려주는 핵심 기능이다.

---

## ✔ 셋의 차이

| 종류 | 특징 | 간단하게 |
| --- | --- | --- |
| EC2 Health Check | 인스턴스 자체만 체크 | 서버 살아있나? |
| ELB Health Check | 서비스 상태까지 체크 (HTTP 응답) | 포트 열려있나? |
| **ALB Health Check** | **타겟 그룹의 경로 체크** | **애플리케이션이 진짜 정상 동작 중인가?** |

---

## 📘 ALB 예시 설정

```yaml
Load Balancer Name: web-alb
Scheme: 인터넷용 (Internet-facing)
Listeners:
   - HTTP : 80 → Target Group
   - HTTPS: 443 → Target Group (ACM SSL 적용)
Target Group:
   Name: web-tg
   Health Check: /health
   Healthy threshold: 2
   Unhealthy threshold: 2
```

# ASG + ALB 통합 아키텍처 예시

## 📌 트래픽 흐름

1️⃣ 사용자가 ALB 주소로 접속

2️⃣ ALB → Target Group → ASG의 EC2로 트래픽 전달

3️⃣ ASG는 Launch Template 기반으로 인스턴스 생성

4️⃣ 부하 증가 시 자동 스케일 아웃

5️⃣ 부하 감소 시 자동 스케일 인

6️⃣ 인스턴스 장애 시 자동 교체

---

## 📘 실제 서비스 구성 예시

```sql
사용자 → ALB → Target Group → ASG → Launch Template → EC2
```

### 예시 상황

- 밤에는 트래픽 적음 → EC2 2대
- 점심시간 트래픽 발생 → CPU 70% → ASG가 2대 추가 → 총 4대
- 2시간 후 트래픽 감소 → 4 → 2대로 자동 축소

# 전체 예제: 웹 서비스 배포 구성 요약

## 1. Launch Template 만들기

- AMI: Amazon Linux 2023
- Instance Type: t3.micro
- User Data: nginx 자동 설치

## 2. ALB 생성

- 리스너: 80 / 443
- Target Group: web-tg (EC2 연결)

## 3. ASG 생성

- Desired: 2
- Min: 2
- Max: 5
- Scaling: CPU 기준 자동 조절
- Target Group 연결

## 4. 결과

- 트래픽 증가하면 서버 자동 증가
- 트래픽 감소하면 자동 감소
- 서버 죽어도 자동 복구
- ALB가 모든 요청을 안정적으로 분배
